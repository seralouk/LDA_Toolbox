```python
import os
import pandas as pd, numpy as np
```

```python
clusters_lev, clusters_fg = np.load("LEV_clusters.npy", allow_pickle=True).tolist(),  \
                                     np.load("FG_clusters.npy", allow_pickle=True).tolist()
```

```python
freesurfer_labels = pd.read_csv('fs_default.txt', delim_whitespace=True, header=None)
names = freesurfer_labels.iloc[:,1].values
```

```python
all_lev = []
for i in range(len(clusters_lev)):
    all_lev.append(names[clusters_lev[i]].tolist())
pd.DataFrame(all_lev).to_csv("LEV_comm_names.csv")
```

```python
all_fg = []
for i in range(len(clusters_fg)):
    all_fg.append(names[clusters_fg[i]].tolist())
pd.DataFrame(all_fg).to_csv("FG_comm_names.csv")
```
